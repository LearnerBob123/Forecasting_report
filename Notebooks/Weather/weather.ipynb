# Weather Data Decomposition

This notebook decomposes a large weather dataset (weather_all.csv) into smaller dataframes grouped by location_id, maps each location to a zone name, and saves one CSV per location under weather_by_zone/.


```python
import pandas as pd

with open("weather_all.csv") as f:
    lines = f.readlines()

start_idx = next(i for i, line in enumerate(lines) if line.startswith("location_id,time"))

df = pd.read_csv("weather_all.csv", skiprows=start_idx)
```

```python
df["time"] = pd.to_datetime(df["time"])
df = df.sort_values(["location_id", "time"])
print(df.shape)
```

```python
grouped = df.groupby("location_id")

for loc_id, group in grouped:
    print(f"Location {loc_id} shape:", group.shape)
```

```python
location_dfs = {loc_id: group.copy()
                for loc_id, group in df.groupby("location_id")}
```

```python
zone_names = [
    "TPNODL_BALASORE",
    "TPNODL_BHADRAK",
    "TPNODL_JAJPUR",
    "TPNODL_BARIPADA",
    "TPNODL_KEONJHAR",
    "TPCODL_ANGUL",
    "TPCODL_PURI",
    "TPCODL_KHORDHA",
    "TPCODL_CUTTACK",
    "TPCODL_JAGATSINGHAPUR",
    "TPCODL_KENDRAPARA",
    "TPCODL_DHENKANAL",
    "TPCODL_NAYAGARH",
    "TPWODL_JHARSUGUDA",
    "TPWODL_SAMBALPUR",
    "TPWODL_ROURKELA",
    "TPWODL_BALANGIR",
    "TPWODL_NUAPADA",
    "TPWODL_BARGARH",
    "TPWODL_KALAHANDI",
    "TPWODL_SONEPUR",
    "TPWODL_DEOGARH",
    "TPSODL_BOUDH",
    "TPSODL_GANJAM",
    "TPSODL_KORAPUT",
    "TPSODL_RAYAGADA",
    "TPSODL_MALKANGIRI",
    "TPSODL_PARALAKHEMUNDI",
    "TPSODL_PHULBANI",
    "TPSODL_NABARANGPUR"
]
```

```python
location_to_zone = dict(enumerate(zone_names))
```

```python
df["zone"] = df["location_id"].map(location_to_zone)
```

```python
import os

os.makedirs("weather_by_zone", exist_ok=True)

for loc_id, sub_df in df.groupby("location_id"):
    zone = location_to_zone[loc_id]
    filename = f"weather_by_zone/weather_{zone}_{loc_id}.csv"
    sub_df.to_csv(filename, index=False)
```

## Note

Workflow summary:
- Read and clean the source file.
- Convert `time` to datetime and sort by `location_id` and `time`.
- Group rows by `location_id`.
- Map each location to a zone name using `location_to_zone`.
- Export each grouped dataframe to `weather_by_zone/weather_<ZONE>_<location_id>.csv`.
